    <h3 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
        <font face="vazir" color="#0099cc">
            بخش اول: پیاده‌سازی توابع مجموعه‌ای
        </font>
    </h3>
    <p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
        <font face="vazir" size=3>
            در این بخش، قصد داریم تمام توابع مربوط به عملیات مجموعه‌ای که در تمرین قبلی معرفی شدند را به صورت تابع‌های مستقل (Functional) پیاده‌سازی کنیم. این توابع شامل عملیات پایه‌ای مانند بررسی خالی بودن مجموعه، بررسی عضویت، زیرمجموعه بودن، محاسبه تعداد اعضای مجموعه توانی، اجتماع، اشتراک، تفاضل و متمم هستند.
            <br>
            شما باید توابع زیر را پیاده‌سازی کنید:
            <ul>
                <li><strong>تابع <code>create_set</code>:</strong> یک لیست را به عنوان ورودی دریافت کرده و آن را به یک مجموعه تبدیل کند.</li>
                <li><strong>تابع <code>is_empty</code>:</strong> بررسی کند که آیا مجموعه تهی است یا خیر.</li>
                <li><strong>تابع <code>is_member</code>:</strong> بررسی کند که آیا عنصر داده‌شده عضو مجموعه است یا خیر.</li>
                <li><strong>تابع <code>is_subset</code>:</strong> بررسی کند که آیا مجموعه اول زیرمجموعه مجموعه دوم است یا خیر.</li>
                <li><strong>تابع <code>power_set_number</code>:</strong> تعداد اعضای مجموعه توانی مجموعه داده‌شده را محاسبه کند.</li>
                <li><strong>تابع <code>union</code>:</strong> اجتماع دو مجموعه را محاسبه کند.</li>
                <li><strong>تابع <code>intersection</code>:</strong> اشتراک دو مجموعه را محاسبه کند.</li>
                <li><strong>تابع <code>difference</code>:</strong> تفاضل دو مجموعه را محاسبه کند.</li>
                <li><strong>تابع <code>complement</code>:</strong> متمم مجموعه نسبت به مجموعه جهانی را محاسبه کند.</li>
            </ul>
            توجه داشته باشید که باید از ساختار داده `set` در پایتون برای پیاده‌سازی این توابع استفاده کنید و هر تابع باید فقط وظیفه مشخصی را انجام دهد. این توابع پایه‌ای هستند و خروجی آن‌ها در بخش‌های بعدی تمرین مورد استفاده قرار خواهد گرفت.
        </font>
    </p>
</div>

```python
def create_set(input_ls):
    return set(input_ls)
    
def union(first_set, second_set):
    return first_set.union(second_set)
    
def intersection(first_set, second_set):
    return first_set.intersection(second_set)
    
def difference(first_set, second_set):
    return first_set.difference(second_set)
    
def complement(universal_set, first_set):
    return universal_set - first_set
    
def is_empty(the_set):
    return len(the_set) == 0
    
def is_member(the_set, element):
    return element in the_set
    
def is_subset(first_set, second_set):
    return first_set.issubset(second_set)
                       
def power_set_number(the_set):
    return 2 ** len(the_set)
```

بخش دوم: پیاده‌سازی توابع محاسبات احتمالاتی
</font>
</h3>
<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
در این بخش، قصد داریم توابع مربوط به محاسبات احتمالاتی را پیاده‌سازی کنیم. این توابع شامل محاسبه احتمال یک پیشامد، احتمال شرطی، بررسی استقلال پیشامدها و اعمال قضیه بیز هستند. این توابع به ما اجازه می‌دهند تا با استفاده از مفاهیم نظریه مجموعه‌ها، محاسبات احتمالاتی را به صورت عملکردی (Functional) پیاده‌سازی کنیم.
<br>
شما باید توابع زیر را پیاده‌سازی کنید:
<ul>
<li>
<strong>تابع <code>probability_of_event</code>:</strong><br>
<em>مثال:</em>  
<code>probability_of_event({1, 2}, {1, 2, 3, 4, 5, 6}) #return 0.333</code><br>
- اولین مجموعه (<code>{1, 2}</code>): مجموعه پیشامد.<br>
- دومین مجموعه (<code>{1, 2, 3, 4, 5, 6}</code>): فضای نمونه (مجموعه جهانی).<br>
- خروجی: 

$$P(A) = \frac{|A|}{|U|} = \frac{2}{6} = 0.333$$


<details class="blue">
<summary>
    <strong>یادآوری مفهوم</strong>
</summary>
احتمال یک پیشامد $ A $ به صورت زیر تعریف می‌شود:

$$P(A) = \frac{|A|}{|U|}$$

که در آن:<br>
- $|A|$ : تعداد عناصر مجموعه پیشامد $ A $.<br>
- $|U|$: تعداد عناصر فضای نمونه (مجموعه جهانی $ U $).<br>
</details>
</li>
<br>
<li>
<strong>تابع <code>conditional_probability</code>:</strong><br>
<em>مثال:</em>  
<code>conditional_probability({1, 2}, {1, 2, 3, 4}) #return 0.5</code><br>
- اولین مجموعه (<code>{1, 2}</code>): مجموعه پیشامد $ A $.<br>
- دومین مجموعه (<code>{1, 2, 3, 4}</code>): مجموعه پیشامد $ B $.<br>
- خروجی: 

$$ P(A|B) = \frac{|A \cap B|}{|B|} = \frac{2}{4} = 0.5 $$

<details class="blue">
<summary>
    <strong>یادآوری مفهوم</strong>
</summary>
احتمال شرطی $P(A|B)$ به صورت زیر تعریف می‌شود:

$$
P(A|B) = \frac{|A \cap B|}{|B|}
$$

که در آن:<br>

- $|A \cap B|$: تعداد عناصر مشترک بین مجموعه‌های $ A $ و $ B $.<br>
- $|B|$: تعداد عناصر مجموعه $ B $.<br>
</details>
</li>
<br>
<li>
<strong>تابع <code>are_independent</code>:</strong><br>
<em>مثال:</em>  
<code>are_independent({1, 2}, {1, 3, 5}, {1, 2, 3, 4, 5, 6}) #retrun True or False</code><br>
- اولین مجموعه (<code>{1, 2}</code>): مجموعه پیشامد $ A $.<br>
- دومین مجموعه (<code>{1, 3, 5}</code>): مجموعه پیشامد $ B $.<br>
- سومین مجموعه (<code>{1, 2, 3, 4, 5, 6}</code>): فضای نمونه (مجموعه جهانی).<br>
- خروجی: True

$$( P(A \cap B) = P(A) \cdot P(B) )$$

<details class="blue">
<summary>
    <strong>یادآوری مفهوم</strong>
</summary>
دو پیشامد $ A $ و $ B $ مستقلند اگر و فقط اگر:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

که در آن:<br>

- $ P(A \cap B) $: احتمال اشتراک دو پیشامد $ A $ و $ B $.<br>
- $ P(A) $: احتمال پیشامد $ A $.<br>
- $ P(B) $: احتمال پیشامد $ B $.<br>

برای بررسی استقلال، ابتدا اشتراک $ A $ و $ B $ را محاسبه کرده و سپس نسبت آن را با حاصلضرب احتمالات $ A $ و $ B $ مقایسه کنید.
</details>
</li>
<br>
<li>
<strong>تابع <code>bayes_theorem</code>:</strong><br>
<em>مثال:</em>  
<code>bayes_theorem({1, 2}, {1, 3, 5}, {1, 2, 3, 4, 5, 6}) #return 0.5</code><br>
- اولین مجموعه (<code>{1, 2}</code>): مجموعه پیشامد $ A $.<br>
- دومین مجموعه (<code>{1, 3, 5}</code>): مجموعه پیشامد $ B $.<br>
- سومین مجموعه (<code>{1, 2, 3, 4, 5, 6}</code>): فضای نمونه (مجموعه جهانی).<br>
- خروجی: 

$$ P(B|A) = \frac{P(A|B) \cdot P(B)}{P(A)} = \frac{\frac{1}{3} \cdot \frac{3}{6}}{\frac{2}{6}} = 0.5 $$

<details class="blue">
<summary>
    <strong>یادآوری مفهوم</strong>
</summary>
قضیه بیز به صورت زیر بیان می‌شود:

$$
P(B|A) = \frac{P(A|B) \cdot P(B)}{P(A)}
$$

که در آن:<br>

- $ P(A|B) $: احتمال شرطی $ A $ به شرط $ B $.<br>
- $ P(B) $: احتمال پیشامد $ B $.<br>
- $ P(A) $: احتمال پیشامد $ A $.<br>

برای محاسبه $ P(B|A) $، ابتدا باید احتمال شرطی $ P(A|B) $ و احتمالات $ P(A) $ و $ P(B) $ را محاسبه کنید.
</details>
</li>
</ul>
توجه داشته باشید که برای پیاده‌سازی این توابع، باید از توابع مجموعه‌ای که در بخش اول پیاده‌سازی کردید، استفاده کنید. به عنوان مثال، برای محاسبه احتمال یک پیشامد، نسبت تعداد عناصر مجموعه پیشامد به تعداد عناصر مجموعه جهانی را محاسبه می‌کنیم. همچنین، برای محاسبه احتمال شرطی و اعمال قضیه بیز، از عملیات مجموعه‌ای مانند اشتراک استفاده می‌کنیم.
</font>
</p>
</div>

```python
def probability_of_event(events, sample_space) -> float:
    return len(events) / len(sample_space)

def conditional_probability(events, sample_space) -> float:
    return len(intersection(events, sample_space)) / len(sample_space)

def are_independent(events1, events2, sample_space) -> float:
    x = probability_of_event(events1, sample_space) * probability_of_event(events2, sample_space)
    y = probability_of_event(intersection(events1, events2), sample_space)
    return abs(x - y) < 1e-9
    

def bayes_theorem(events1, events2, sample_space) -> float:
    x = conditional_probability(events1, events2) * probability_of_event(events2, sample_space)
    return x / probability_of_event(events1, sample_space)

```

در این بخش، قصد داریم توابع پیاده‌سازی‌شده را با استفاده از مثال‌های مختلف تست کنیم و نتایج حاصل از آن‌ها را بررسی کنیم. هدف از این مرحله، تأیید درست‌کاری توابع و بررسی صحت محاسبات است.
<br>
برای تست توابع، می‌توانید از مجموعه‌های مختلف و مسائل واقعی استفاده کنید. در ادامه، چند مثال تست برای هر تابع ارائه شده است:
<ul>
<li>
<strong>تابع <code>probability_of_event</code>:</strong><br>
<em>مثال تست:</em>  
<code>probability_of_event({1, 2, 3}, {1, 2, 3, 4, 5, 6})</code><br>
- اولین مجموعه (<code>{1, 2, 3}</code>): مجموعه پیشامد.<br>
- دومین مجموعه (<code>{1, 2, 3, 4, 5, 6}</code>): فضای نمونه (مجموعه جهانی).<br>
- خروجی: 

$$ P(A) = \frac{|A|}{|U|} = \frac{3}{6} = 0.5 $$

<details class="blue">
<summary>
<strong>راهنمایی برای تست</strong>
</summary>
برای تست این تابع، می‌توانید مجموعه‌های مختلفی با اندازه‌های متغیر امتحان کنید. مثلاً:<br>
- فضای نمونه: <code>{1, 2, 3, 4, 5, 6, 7, 8}</code>.<br>
- مجموعه پیشامد: <code>{2, 4, 6}</code>.<br>
- نتیجه منتظره: 

$$ P(A) = \frac{3}{8} = 0.375 $$

اطمینان حاصل کنید که خروجی تابع با محاسبات دستی سازگار است.
</details>
</li>
<br>
<li>
<strong>تابع <code>conditional_probability</code>:</strong><br>
<em>مثال تست:</em>  
<code>conditional_probability({1, 2}, {1, 2, 3, 4})</code><br>
- اولین مجموعه (<code>{1, 2}</code>): مجموعه پیشامد $ A $.<br>
- دومین مجموعه (<code>{1, 2, 3, 4}</code>): مجموعه پیشامد $ B $.<br>
- خروجی: 

$$ P(A|B) = \frac{|A \cap B|}{|B|} = \frac{2}{4} = 0.5 $$

<details class="blue">
<summary>
<strong>راهنمایی برای تست</strong>
</summary>
برای تست این تابع، می‌توانید مجموعه‌های مختلفی با اشتراک‌های متغیر امتحان کنید. مثلاً:<br>

- مجموعه $ A $: <code>{2, 4, 6}</code>.<br>
- مجموعه $ B $: <code>{1, 2, 3, 4}</code>.<br>
- نتیجه منتظره: $ P(A|B) = \frac{|A \cap B|}{|B|} = \frac{2}{4} = 0.5 $

اطمینان حاصل کنید که تابع به درستی اشتراک دو مجموعه را محاسبه کرده و نسبت را به درستی برمی‌گرداند.
</details>
</li>
<br>
<li>
<strong>تابع <code>are_independent</code>:</strong><br>
<em>مثال تست:</em>  
<code>are_independent({1, 2}, {3, 4}, {1, 2, 3, 4, 5, 6})</code><br>
- اولین مجموعه (<code>{1, 2}</code>): مجموعه پیشامد $ A $.<br>
- دومین مجموعه (<code>{3, 4}</code>): مجموعه پیشامد $ B $.<br>
- سومین مجموعه (<code>{1, 2, 3, 4, 5, 6}</code>): فضای نمونه (مجموعه جهانی).<br>
- خروجی: True چون 

$$ P(A \cap B) = P(A) \cdot P(B) $$

<details class="blue">
<summary>
<strong>راهنمایی برای تست</strong>
</summary>
برای تست این تابع، می‌توانید مجموعه‌های مستقل و وابسته را بررسی کنید. مثلاً:<br>

- مجموعه $ A $: <code>{1, 2}</code>.<br>
- مجموعه $ B $: <code>{3, 4}</code>.<br>
- فضای نمونه: <code>{1, 2, 3, 4, 5, 6}</code>.<br>
- نتیجه منتظره: 
True (چون اشتراک $ A $ و $ B $ تهی است و $ P(A \cap B) = P(A) \cdot P(B) $).<br>

همچنین، می‌توانید مجموعه‌های وابسته را نیز تست کنید تا اطمینان حاصل کنید که تابع نتیجه صحیح را برمی‌گرداند.
</details>
</li>
<br>
<li>
<strong>تابع <code>bayes_theorem</code>:</strong><br>
<em>مثال تست:</em>  
<code>bayes_theorem({1, 2}, {1, 3, 5}, {1, 2, 3, 4, 5, 6})</code><br>
- اولین مجموعه (<code>{1, 2}</code>): مجموعه پیشامد $ A $.<br>
- دومین مجموعه (<code>{1, 3, 5}</code>): مجموعه پیشامد $ B $.<br>
- سومین مجموعه (<code>{1, 2, 3, 4, 5, 6}</code>): فضای نمونه (مجموعه جهانی).<br>
- خروجی: 

$$ P(B|A) = \frac{P(A|B) \cdot P(B)}{P(A)} = \frac{\frac{1}{3} \cdot \frac{3}{6}}{\frac{2}{6}} = 0.5 $$

- فضای نمونه: <code>{1, 2, 3, 4, 5, 6}</code>.<br>
- نتیجه منتظره: 

$$ P(B|A) = 0.5 $$
